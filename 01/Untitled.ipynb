```python
import pandas as pd
```

```python
depths = pd.read_csv('readings.csv', header=None)
depths
```

```python
depths_list = depths.values.tolist()
print(depths_list)
```

```python
# Count
increase_count = 0

# zip the two lists together so you compare list 1 to list 2 + 1 index position forward
for x, y in zip(depths_list, depths_list[1:]):
    if x < y:
        increase_count +=1 # add to count if it's higher
        print(x,y) # check work to see if it's comparing correctly
    else: 
        pass
```
